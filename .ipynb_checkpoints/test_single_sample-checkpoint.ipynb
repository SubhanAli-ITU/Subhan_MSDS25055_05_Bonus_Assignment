This notebook loads the trained DDPM model and demonstrates:
1. Generating a grid of new images
2. Visualising the denoising trajectory (noise → image)
3. Visualising the forward noising process at selected time-steps

```python
import os, torch, numpy as np, matplotlib.pyplot as plt
import torchvision.utils as vutils
from IPython.display import display
from PIL import Image

# ── Import project modules ──────────────────────────────────
from MSDS25055_05 import UNet, NoiseScheduler, sample, denorm, AnimalDataset
from MSDS25055_05_test import sample_with_trajectory

DEVICE     = 'cuda' if torch.cuda.is_available() else 'cpu'
MODEL_PATH = 'saved_models/ddpm_final.pth'  # ← change if needed
IMAGE_SIZE = 64
T          = 1000
BASE_CH    = 64

print(f'Device: {DEVICE}')
```

```python
# ── Load model ──────────────────────────────────────────────
ckpt = torch.load(MODEL_PATH, map_location=DEVICE)
state_dict = ckpt['model_state'] if 'model_state' in ckpt else ckpt

model = UNet(in_ch=3, base_ch=BASE_CH, ch_mults=(1,2,4), time_emb_dim=256).to(DEVICE)
model.load_state_dict(state_dict)
model.eval()

scheduler = NoiseScheduler(T=T, device=DEVICE)
print('Model loaded successfully.')
```

## 1. Generate a grid of 8 new images

```python
gen = sample(model, scheduler, n_samples=8, image_size=IMAGE_SIZE, device=DEVICE)

grid = vutils.make_grid(denorm(gen), nrow=4, padding=2)
plt.figure(figsize=(12, 6))
plt.imshow(grid.permute(1,2,0).cpu().numpy())
plt.title('Generated Images (DDPM) – MSDS25055', fontsize=14)
plt.axis('off')
plt.tight_layout()
plt.savefig('generated_grid.png', dpi=150)
plt.show()
```

## 2. Denoising trajectory (Noise → Image)

```python
final, traj = sample_with_trajectory(
    model, scheduler, image_size=IMAGE_SIZE, device=DEVICE, n_snapshots=10
)

n_cols = len(traj) + 1
fig, axes = plt.subplots(1, n_cols, figsize=(2.2*n_cols, 2.5))
fig.suptitle('Denoising Trajectory  (x_T → x_0)  –  MSDS25055', fontsize=11)

for i, img_t in enumerate(traj):
    step = T - 1 - int(i * (T-1) / max(len(traj)-1, 1))
    axes[i].imshow(denorm(img_t[0]).permute(1,2,0).numpy())
    axes[i].set_title(f't={step}', fontsize=7)
    axes[i].axis('off')

axes[-1].imshow(denorm(final[0]).permute(1,2,0).numpy())
axes[-1].set_title('Final (x₀)', fontsize=7)
axes[-1].axis('off')

plt.tight_layout()
plt.savefig('denoising_trajectory.png', dpi=150, bbox_inches='tight')
plt.show()
```

## 3. Forward noising process visualisation

```python
# Requires dataset path — update accordingly
DATASET_PATH = 'dataset'   # ← set your dataset path here

if os.path.exists(DATASET_PATH):
    ds     = AnimalDataset(DATASET_PATH, image_size=IMAGE_SIZE)
    x0     = ds[0].unsqueeze(0).to(DEVICE)
    steps  = torch.linspace(0, T-1, 11, dtype=torch.long)

    noisy_imgs = [denorm(x0).cpu()]
    for t_val in steps:
        xt, _ = scheduler.add_noise(x0, torch.tensor([t_val], device=DEVICE))
        noisy_imgs.append(denorm(xt).cpu())

    fig, axes = plt.subplots(1, len(noisy_imgs), figsize=(2.2*len(noisy_imgs), 2.5))
    titles    = ['Original'] + [f't={s.item()}' for s in steps]
    for ax, img, title in zip(axes, noisy_imgs, titles):
        ax.imshow(img[0].permute(1,2,0).numpy())
        ax.set_title(title, fontsize=7)
        ax.axis('off')
    fig.suptitle('Forward Diffusion  (x_0 → x_T)  –  MSDS25055', fontsize=11)
    plt.tight_layout()
    plt.savefig('forward_diffusion.png', dpi=150, bbox_inches='tight')
    plt.show()
else:
    print(f'Dataset not found at {DATASET_PATH!r}. Skipping forward vis.')
```
